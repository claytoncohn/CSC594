```
import pandas as pd

DATA_PATH = "drive/My Drive/colab/data/"
DATA_NAME = "EBA1415-SkinCancer-big-sentences.tsv"
big = pd.read_csv(DATA_PATH + DATA_NAME, delimiter='\t', header=0, names=['file', 'relation', 's_num', 'sentence'], index_col = False)
big.head(85)
```

```
sents = big["sentence"]
sents.head()
```

```
bad = []
for index, value in sents.items():
    if "\n" in value:
      bad.append(index)
print(bad)
```

```
big.drop(bad, inplace=True)
sents = big["sentence"]
sents.head()
```

```
bad = []
for index, value in sents.items():
    if "\n" in value:
      bad.append(index)
print(bad)
```

```
big.head(85)
```

```
DATA_NAME = "EBA1415-SkinCancer-little-sentences.tsv"
little = pd.read_csv(DATA_PATH + DATA_NAME, delimiter='\t', header=0, names=['file', 'relation', 's_num', 'sentence'])
little.head(100)
```

```
sents = little["sentence"]
sents.head()
```

```
bad = []
for index, value in sents.items():
    if "\n" in value:
      bad.append(index)
print(bad)
```

```
little.drop(bad, inplace=True)
sents = little["sentence"]
sents.head()
```

```
bad = []
for index, value in sents.items():
    if "\n" in value:
      bad.append(index)
print(bad)
```
