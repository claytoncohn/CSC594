```
import pandas as pd

DATA_PATH = "drive/My Drive/colab/data/"
DATA_NAME = "skin_cancer_all.tsv"
df = pd.read_csv(DATA_PATH + DATA_NAME, delimiter='\t', header=0, names=['file', 'relation', 's_num', 'sentence'])
```

```
print(df.shape)
df.head(10)
```

```
d = {}
for index, row in df.iterrows():
  r = row.relation
  if r in d:
    val = d[r]
    val.append(index)
    d[r] = val
  else:
    d[r] = [str(index)]
```

```
def print_dict(dct):
  for key, value in dct.items():
    print(key, value)
```

```
print_dict(d)
```

```
sorted_list = sorted(d.items(), key= lambda value: len(value[1]), reverse=True)
print(sorted_list)
```

```
d_sorted = {x[0]:x[1] for x in sorted_list}
print_dict(d_sorted)
```

```
d_iso = {}
for k, v in d_sorted.items():
  if len(v) >= 15:
    d_iso[k] = v
```

```
print_dict(d_iso)
```

```
print("Classes:",len(d_iso))
print("Dist:")
for k, v in d_iso.items():
  print("\t{}:\t\t{}".format(k, len(v)))
```

```
import random

for k, v in d_iso.items():
  d_iso[k] = random.sample(v, k=15)
```

```
print_dict(d_iso)
```

```
print("Iter: 2")
print("Classes:",len(d_iso))
print("Dist:")
for k, v in d_iso.items():
  print("\t{}:\t\t{}".format(k, len(v)))
```

```
d_final = d_iso
```

```
values = []
for k, v in d_final.items():
  values += v
values = [int(i) for i in values]
values.sort()
print(values)
```

```
df_filtered = df.iloc[values]
```

```
df_filtered.head(10)
```

```
df_filtered.to_csv(DATA_PATH + "skin_cancer_2.tsv", sep='\t', index=False)
```
